 # FEATURE SCALING

## MIN_MAX_SCALER


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
df=pd.read_csv("train.csv",usecols=["Pclass","Survived","Age","Fare"])
df.head()
```

```python
norm=MinMaxScaler()
df.Age.fillna(df.Age.median(),inplace=True)
```

```python
data=pd.DataFrame(norm.fit_transform(df),columns=["Survived","Pclass","Age","Fare"])
```

```python
data
```

## ROBUST SCALER

```python
df2=pd.read_csv("train.csv",usecols=["Pclass","Survived","Age","Fare"])
df2.Age.fillna(df.Age.median(),inplace=True)
df2.head()
```
