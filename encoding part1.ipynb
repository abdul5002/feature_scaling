## FREQUENCY OR COUNT ENCODING(NOMINAL)

```python
data1=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None)
data1.head()
```

```python
columns=[1,3,5,6,7,8,9,13]
data1=data1[columns]
```

```python
data1.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
data1.head()
```

```python
for column in data1.columns:
    print(column,":",len(data1[column].unique()))
```

```python
country=data1.Country.value_counts().to_dict()
```

```python
data1["encoded_country"]=data1.Country.map(country)
```
