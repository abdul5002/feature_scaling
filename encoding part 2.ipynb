# ENCODING

## TARGET GUIDED ORDINAL ENCODING(ORDINAL)

```python
import pandas as pd
df=pd.read_csv("train.csv",usecols=['Cabin','Embarked','Survived'])
df.head()
```

```python
df.Cabin.fillna("Missing",inplace=True)
df.groupby('Survived')['Cabin']
```

```python
df['Cabin']=df.Cabin.astype(str).str[0]
```

```python
df
```

```python
df.Cabin.unique()
```

```python
labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ord1={k:i for i,k in enumerate(labels,0)}
```

```python
ord1
```

```python
df['cabin']=df.Cabin.map(ord1)
```

```python
df
```

## PROBABILITY RATIO ENCODING

```python
import pandas as pd
df3=pd.read_csv("train.csv",usecols=['Cabin','Embarked','Survived'])
df3.head()
```

```python
df3.Cabin.fillna("Missing",inplace=True)
df3['Cabin']=df3.Cabin.astype(str).str[0]
```

```python
df3
```

```python
survived=df3.groupby(['Cabin'])['Survived'].mean()
```

```python
survived
```

```python
Not_survived=1-survived
```

```python
Not_survived
```

```python
prob=(survived/Not_survived).to_dict()
```

```python
prob
```

```python
df3['Cabin']=df3.Cabin.map(prob)
```

```python
df3
```
